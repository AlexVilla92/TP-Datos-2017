```python
#Cambio de grados Fahrenheit a grados Celcius
weather['max_temperature_f'] = weather['max_temperature_f'].map(lambda x: (x-32)/1.8)
weather['min_temperature_f'] = weather['min_temperature_f'].map(lambda x: (x-32)/1.8)
weather['mean_temperature_f'] = weather['mean_temperature_f'].map(lambda x: (x-32)/1.8)
```
